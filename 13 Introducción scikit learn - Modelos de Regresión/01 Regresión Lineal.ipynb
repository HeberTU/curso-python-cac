## Qu es regresin?

**La regresin busca relaciones entre variables.**

Por ejemplo, los salarios dependen de caractersticas como:

- Experiencia
- Nivel de educacin
- Rol
- Ciudad, 
- etc.

La presuncin es que la experiencia, la educacin, el rol y la ciudad son caractersticas independientes, mientras que el salario depende de ellas.


Siguiendo el supuesto de que al menos una de las caractersticas depende de las dems, se intenta establecer una relacin entre ellas.

>De manera ms formal, buscamos una funcin que mapee algunas variables a otra lo suficientemente bien.

Las caractersticas dependientes se denominan:

- variables dependientes
- Outputs
- Respuestas

Las caractersticas independientes se denominan:

- variables independientes
- Inputs
- Regresores
- Predictores.

Los problemas de regresin tienen una variable dependiente numrica. Sin embargo, los regresores pueden ser datos continuos, discretos o incluso categricos, como gnero, nacionalidad o provincia.

Es una prctica comn denotar a la variable respuesta como $y$ y los regresores como $x$.

Si hay dos o ms regresores, se pueden representar como el vector **x** $ = (x_1, x_2, x_3, ... x_r)$, donde $r$ es el nmero de observaciones.

## Cundo necesitamos una regresin?

Por lo general podemos usarla para dos cosas:

1. **Inferencia**: Para determinar el grado de dependencia de la variable respuesta y uno o ms regresores. En este casos formulamos una hiptesis y la comprobamos con los datos.


2. **Pronosticar**: Si para una nueva observacin slo tenemos los regresores y no la variable respuesta podemos utilizar los regresores para estimar la variable respuesta.

## Regresin Lineal

Es probablemente una de las tcnicas de regresin ms importantes y ampliamente utilizadas. Es uno de los mtodos de regresin ms simples. Una de sus principales ventajas es la facilidad de interpretacin de los resultados.


### Formulacin del problema

Dado:

- $y$ la variable respuesta.
- **x**$ = (x_1, x_2, ...,x_r)$ el conjunto re regresores.
- $r$ El nmero de regresores

Asumiremos una relacin lineal entre $y$ y **x** tal que:

$$ y = \beta_0 + \beta_1x_1 + \beta_2x_2 + ... + \beta_rx_r + \epsilon$$

Donde

- $\beta_0, \beta_1, \beta_2, ..., \beta_r$ son los coeficientes de regresin
- $\epsilon$ es el error aleatorio.

La regresin lineal estimar los coeficientes de regresin, denotados como , , , . Estos estimadores definen la funcin de regresin estimada:

$$ f(x) = b_0 + b_1x_1 + b_2x_2 + ... + b_rx_r $$

Esta funcin debera capturar las dependencias entre los regresores **x** y la variable respuesta $y$ suficientemente bien.

Es decir $f(x_i)$ debe estar lo ms cerca posible de $y_i$.

Las diferencias $y_i - f(x_i)$ se llaman residuos.

La regresin consiste en encontrar los mejores valores de , , ,  tal que los residuos sean lo ms posible pequeos.

Para obtener los mejores pesos, generalmente minimizamos la suma de los residuos cuadrados 

$$ SSR = \sum \limits_{i=1} ^{n} (y_i - f(x_i))^2$$

> Este enfoque se llama el mtodo de los mnimos cuadrados ordinarios.

## Performance 

La varianza de la variable respuesta $y_i$, $i = 1, 2, ..., n$ se debe en parte a la dependencia de los regresores $x_i$. Sin embargo, tambin hay una variacin inherente adicional:

- Error de precisin en la medicin.
<br>

- Dependencia de otras variables que no han sido registradas.


El **coeficiente de determinacin**, denotado como $R^2$, nos indica la cantidad de variacin en $y$ que se puede explicar por la dependencia de $x$, usando el modelo de regresin particular. 


- Un $R^2 \epsilon [0, 1]$

- Un $R^2$ ms grande indica un mejor ajuste

### Regresin lineal

Por el momento slo exploraremos las funcionalidades de la clase `LinearRegression`.

#### Paso 1: Importamos las dependencias

```python
import numpy as np
from sklearn.linear_model import LinearRegression
```

#### Paso 2: Generemos algunos datos:

```python
x = np.array([5, 15, 25, 35, 45, 55]).reshape((-1, 1))
y = np.array([5, 20, 14, 32, 22, 38])
```

Hemos usado el mtodo `reshape()` para hacer un `array` de dos dimensones, i.e una columna y 6 renglones.

La tupla `(-1, 1)` Se usa para definir el nuevo formato, la primera entrada representa el nmero de renglones y la segunda el nmero de columnas. El valor `-1` indica an numpy crear tantos renglones como sea posible.

#### Paso 3: Instanciar un objeto de la clase `LinearRegression`

Podemos especificar parmetros opcionales como:

- `fit_intercept`: `True` si queremos estimar el valor $b_0$.

- `normalize`: `True` si queremos normalizar los datos de entrada.

- `n_jobs`: El nmero de hilos de proceso en paralelo para ajustar el modelo. 

```python
model = LinearRegression()
```

#### Paso 4: Ajustar el modelo.

El mtodo `.fit()` calcula los valores optimo de $b_0$ y $b_1$ dados los datos.

```python
model.fit(x, y)
```

#### Paso 5: Obtener resultados.

Podemos obtener el coeficiente de determinacin con el mtodo `.score()`

```python
r_sq = model.score(x, y)
print(f"coefficiente de determinacin: {r_sq}")
```

Podemos obtener los valores de $b_0 y b_1$

```python
print(f"b0: {model.intercept_}")
print(f"b1: {model.coef_}")
```

Podemos usar el mtodo `.predict()` para hacer inferencia sobre nuevas observaciones.

```python
x_new = np.array([10]).reshape((-1,1))
y_pred = model.predict(x_new)
y_pred
```

### Regresin polinmica


Es posible transformar la matriz de entrada de varias maneras, e.g. usar `np.insert()`. Pero la clase `PolynomialFeatures` nos ayuda a crear un `pipeline` ms limpio.


```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

x = np.array([5, 15, 25, 35, 45, 55]).reshape((-1, 1))
y = np.array([5, 20, 14, 32, 22, 38])

transformer = PolynomialFeatures(degree=2, include_bias=False)
transformer.fit(x)
x_ = transformer.transform(x)

model = LinearRegression().fit(x_, y)


r_sq = model.score(x_, y)
print(f"coeficiente de determinacin: {r_sq}")

print(f"b0: {model.intercept_}")


print(f"b1 y b2: {model.coef_}")
```

Usando la clase `PolynomialFeatures` podemos genear predicciones de manera ms sencilla.

```python
x_new = np.array([10]).reshape((-1,1))
transformer.transform(x_new)
```

```python
model.predict(transformer.transform(x_new))
```
